```python
import pandas as pd
import numpy as np

from paths import *
from analysis_utils import get_aggregated_dataframe

# create plot output folders
import os
plot_output_folder = f"test_results_plots"
os.makedirs(plot_output_folder, exist_ok=True)


df_results_aggregated, df_case, df_case_merged = get_aggregated_dataframe(experiment_folders_test)
df_case_merged
```

```python
import itertools

df_case_merged["prompt_name"] = df_case_merged["System Prompt"] + " + " + df_case_merged["User Prompt"]

print(f"Found {len(df_case_merged['prompt_name'].value_counts())} prompts:", df_case_merged["prompt_name"].unique())

list_of_result_dicts = []

list_of_result_dicts2 = []

for model in df_case_merged["model_name"].unique():
    df_model = df_case_merged[df_case_merged["model_name"] == model]
    prompt_pairs = list(itertools.combinations(df_model["prompt_name"].unique(), 2))
    print(f"\n=== Model: {model} ===")
    print(f"Prompt pairs: {prompt_pairs}")

    # for each prompt pair
    scores_per_prompt = []
    for prompt1, prompt2 in prompt_pairs:
        df_prompt1 = df_model[df_model["prompt_name"] == prompt1]
        df_prompt2 = df_model[df_model["prompt_name"] == prompt2]

        # merge on idx_new to compare predictions for the same cases
        is_correct_1 = df_prompt1.set_index("idx_new")["is_correct_maj"].astype(int)
        is_correct_2 = df_prompt2.set_index("idx_new")["is_correct_maj"].astype(int)
        
        score = np.abs(is_correct_1 - is_correct_2)
        scores_per_prompt.append(score)
    scores_per_prompt = np.array(scores_per_prompt)
    avg_score_per_sample = scores_per_prompt.mean(axis=0)
    assert avg_score_per_sample.shape[0] == 200, f"Expected 200 samples, but got {avg_score_per_sample.shape[0]}"

    sensitivity_score = avg_score_per_sample.mean()
    print(f"Average sensitivity score across prompt pairs: {sensitivity_score:.4f}")

    list_of_result_dicts.append({
        "model_name": model.split("/")[-1],
        "sensitivity_score": sensitivity_score
    })

    if "SP5 + UP2" in df_model["prompt_name"].unique():
        list_of_result_dicts2.append({
            "model_name": f'{model.split("/")[-1]}\n(Prompt sensitivity)',
            "sensitivity_score": sensitivity_score
        })

print("\n=== Prompt Sensitivity Scores ===")
print(f"Min sensitivity score: {min([d['sensitivity_score'] for d in list_of_result_dicts]):.4f}")
print(f"Max sensitivity score: {max([d['sensitivity_score'] for d in list_of_result_dicts]):.4f}")

df_prompr_sensitivity = pd.DataFrame(list_of_result_dicts)
import plotnine as p9
plt = (p9.ggplot(df_prompr_sensitivity, p9.aes(x="reorder(model_name, sensitivity_score)", y="sensitivity_score", fill="sensitivity_score")) + 
       p9.scale_fill_gradient(low='#2ecc71', high='#e74c3c', name="PromptSensiScore", limits=(0.03, 0.11)) +
       p9.geom_bar(stat="identity") + 
       p9.ylab("PromptSensiScore") +
       p9.xlab("") +
       # x axis to log
       p9.coord_flip() +
    
       p9.theme_bw() + 
       p9.theme(figure_size=(9,4), axis_text= p9.element_text(size=12), legend_position="right")
)
plt.show()
plt.save("test_results_plots/prompt_sensitivity_scores.pdf", dpi=300)

for prompt in df_case_merged["prompt_name"].unique():
    if prompt != "OPTION_6b + OPTION_2":
        continue
    df_model = df_case_merged[df_case_merged["prompt_name"] == prompt]
    prompt_pairs = list(itertools.combinations(df_model["model_name"].unique(), 2))
    print(f"\n=== Prompt: {prompt} ===")
    print(f"Prompt pairs: {prompt_pairs}")

    # for each prompt pair
    scores_per_prompt = []
    for prompt1, prompt2 in prompt_pairs:
        df_prompt1 = df_model[df_model["model_name"] == prompt1]
        df_prompt2 = df_model[df_model["model_name"] == prompt2]

        # merge on idx_new to compare predictions for the same cases
        is_correct_1 = df_prompt1.set_index("idx_new")["is_correct_maj"].astype(int)
        is_correct_2 = df_prompt2.set_index("idx_new")["is_correct_maj"].astype(int)
        
        score = np.abs(is_correct_1 - is_correct_2)
        scores_per_prompt.append(score)
    scores_per_prompt = np.array(scores_per_prompt)
    avg_score_per_sample = scores_per_prompt.mean(axis=0)
    assert avg_score_per_sample.shape[0] == 200, f"Expected 200 samples, but got {avg_score_per_sample.shape[0]}"

    sensitivity_score = avg_score_per_sample.mean()
    print(f"Average sensitivity score across prompt pairs: {sensitivity_score:.4f}")

    list_of_result_dicts2.append({
        "model_name": "SP5 + UP2\n(Model sensitivity)",
        "sensitivity_score": sensitivity_score
    })

df_prompr_sensitivity2 = pd.DataFrame(list_of_result_dicts2)
#" print min and max sensitivity scores: "
print(f"Min sensitivity score: {df_prompr_sensitivity2['sensitivity_score'].min():.4f}")
print(f"Max sensitivity score: {df_prompr_sensitivity2['sensitivity_score'].max():.4f}")
import plotnine as p9
plt = (p9.ggplot(df_prompr_sensitivity2, p9.aes(x="reorder(model_name, sensitivity_score)", y="sensitivity_score", fill="sensitivity_score")) + 
       p9.scale_fill_gradient(low='#2ecc71', high='#e74c3c', name="PromptSensiScore", limits=(0.03, 0.11)) +
       p9.geom_bar(stat="identity") + 
       p9.ylab("PromptSensiScore") +
       p9.xlab("") +
       # x axis to log
       p9.coord_flip() +
    
       p9.theme_bw() + 
       p9.theme(figure_size=(9,4), axis_text= p9.element_text(size=12), legend_position="right")
)
plt.show()
plt.save("test_results_plots/prompt_sensitivity_scores_combined.pdf", dpi=300)
```
